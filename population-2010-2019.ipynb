### Importing pandas

```python
import pandas as pd
```

#### getting data from csv

```python
population2010_2019 = pd.read_csv('./population_data/population-2010-2019.csv')
population2010_2019.head()
```

#### we need to remove top row and make header correct

```python

new_header = population2010_2019.iloc[0] 
df_population2010_2019 = population2010_2019[1:] 
df_population2010_2019.columns = new_header 

df_population2010_2019.head()
```

#### Renaming column name Time to Year


```python
df_population2010_2019.rename(columns={'Time': 'Year'}, inplace=True)
df_population2010_2019.head()
```

```python
df_population2010_2019.describe()
```

```python
df_population2010_2019.info()   
```

#### Sort by 'AgeGrp' in descending order

```python
sorted_df_population2010_2019 = df_population2010_2019.sort_values(by='AgeGrp', ascending=False)
sorted_df_population2010_2019
```

#### Correcting values by Replacing values

```python
df_population2010_2019['AgeGrp'] = df_population2010_2019['AgeGrp'].replace({'Okt 14': '10-14', '05. Sep': '5-9','905. Sep9':'9-9', '100+':'100-104', '2982616.38':'5-9'})
df_population2010_2019
```

```python
sorted_df2_population2010_2019 = df_population2010_2019.sort_values(by='AgeGrp', ascending=False)
sorted_df2_population2010_2019
```

#### Removing unwanted column

```python
df_population2010_2019.drop(columns=['AgeGrpSpan'], inplace=True)
df_population2010_2019
```

```python
sorted_df4_population2010_2019 = df_population2010_2019.sort_values(by='AgeGrp', ascending=True)
sorted_df4_population2010_2019
```

#### Removig Null Values

```python
df_population2010_2019.dropna(inplace=True)
df_population2010_2019
```

```python
sorted_df3_population2010_2019 = df_population2010_2019.sort_values(by='AgeGrp', ascending=True)
sorted_df3_population2010_2019
```
